## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Remove Duplicates
df.drop_duplicates(inplace=True)

# Handle Missing Values
# drop rows where 'Salary Estimate' is missing
df.dropna(subset=['Salary Estimate'], inplace=True)

# fill missing company ratings with the mean rating
mean_rating = df['Rating'].mean()
df['Rating'].fillna(mean_rating, inplace=True)

# fill missing company sizes with 'Unknown'
df['Size'].fillna('Unknown', inplace=True)

#Clean 'Salary Estimate' Column Remove extraneous text
df['Salary Estimate'] = df['Salary Estimate'].str.replace('Glassdoor est.', '', regex=False)
df['Salary Estimate'] = df['Salary Estimate'].str.replace('Employer Provided Salary:', '', regex=False)
df['Salary Estimate'] = df['Salary Estimate'].str.replace('per hour', '', case=False, regex=False)

# Extract Numeric Salary Ranges
# Create min_salary, max_salary, and a computed average
salary_split = df['Salary Estimate'].str.split('-', expand=True)

# Remove non-digit characters (e.g., $, K, etc.)
salary_split[0] = salary_split[0].str.replace(r'\D', '', regex=True)
salary_split[1] = salary_split[1].str.replace(r'\D', '', regex=True)

df['min_salary'] = pd.to_numeric(salary_split[0], errors='coerce')
df['max_salary'] = pd.to_numeric(salary_split[1], errors='coerce')
df['avg_salary_calculated'] = (df['min_salary'] + df['max_salary']) / 2

# Clean 'Company Name' (Remove appended rating text in parentheses, if present)
def clean_company_name(name):
  return name.split('(')[0].strip() if '(' in name else name.strip()

df['Company Name'] = df['Company Name'].apply(clean_company_name)

# Extract State from 'Location' (If 'Location' is in "City, State" format)

df['job_state'] = df['Location'].apply(
    lambda x: x.split(',')[-1].strip() if pd.notnull(x) and ',' in x else x
)

# Convert Certain Columns to Categorical (if appropriate)
for cat_col in ['Size', 'Type of ownership', 'Industry', 'Sector']:
  if cat_col in df.columns:
    df[cat_col] = df[cat_col].astype('category')

# Filter Out Unrealistic Salaries (Example: remove rows with min_salary <= 0 or max_salary extremely high)
df = df[(df['min_salary'] > 0) & (df['max_salary'] < 1000)]

# EXAMPLE ANALYSIS

## 9.1 Average Salary by Job Title
if 'avg_salary' not in df.columns:
    df['avg_salary'] = df['avg_salary_calculated']
job_title_salary = df.groupby('Job Title')['avg_salary'].mean().sort_values(ascending=False)
print("Average Salary by Job Title (Top 5):")
print(job_title_salary.head(5), "\n")

## 9.2 Distribution of 'avg_salary'
# (This uses the dataset's original 'avg_salary' column, if already present.
#  You could also compare it to 'avg_salary_calculated' you created above.)
sns.histplot(df['avg_salary'], kde=True, bins=30)
plt.title("Distribution of Average Salary")
plt.xlabel("Average Salary")
plt.show()

## 9.3 Relationship Between Rating & Average Salary
sns.scatterplot(data=df, x='Rating', y='avg_salary')
plt.title("Company Rating vs. Average Salary")
plt.show()

## 9.4 Correlation Matrix (Numeric Columns)
# Candidate numeric columns list as per the original code
candidate_numeric_cols = ['Rating', 'avg_salary', 'num_comp', 'sdesc_len', 'avg_salary_calculated', 'min_salary', 'max_salary']

# Filter the list to include only columns present in the DataFrame
numeric_cols = [col for col in candidate_numeric_cols if col in df.columns]

# Calculate the correlation matrix for the available numeric columns
corr_matrix = df[numeric_cols].corr()

# Plot the correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
````

## Manipulations

**Cleaned the salary estimates:** Removed extraneous text (such as "Glassdoor est." and "per hour") and split the salary range into minimum and maximum values, then computed an average salary.

**Cleaned company names:** Stripped out any appended ratings or extra characters from the company name.

**Converted data types:** Cast columns like company size, industry, and sector to categorical types for more efficient analysis.

**Filtered out outliers:** Removed records with unrealistic salary values to ensure the analysis focused on valid entries.
